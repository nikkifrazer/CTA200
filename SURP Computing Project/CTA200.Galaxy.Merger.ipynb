```python
#Part 3: Plotting Galaxy Merger Trees
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 

#Question 2
#read in the CSV data using Pandas
data = pd.read_csv("C:/Users/nicol/Uni 2019-2020/Python 2020/G2.Galaxy.Data.csv", delimiter = ',')
```

```python
#Question 3: Scatter Plots 
#Plots 1: Galaxy ID vs. redshift
plt.scatter(data['galaxyID'], data['redshift'], marker='.', color='red')
plt.xlabel("Galaxy ID")
plt.ylabel("Redshift")
plt.title("Galaxy ID vs. Redshift")
plt.savefig("CTA200.GalaxyIDperz.pdf")
plt.show()
```

```python
#Plots 2: stellar mass per redshift
plt.scatter(data['stellarMass'], data['redshift'], marker='.', color='firebrick')
plt.xlabel('Stellar Mass')
plt.ylabel("Redshift")
plt.title('Stellar Mass per Redshift')
plt.savefig("CTA200.stellarmassperz.pdf")
plt.show()
```

```python
#Plots 3: virial mass per redshift
plt.scatter(data['mvir'], data['redshift'], marker='.', color='maroon')
plt.xlabel('Virial Mass')
plt.ylabel("Redshift")
plt.title('Virial Mass per Redshift')
plt.savefig("CTA200.mvirperz.pdf")
plt.show()
```

```python
#Plots 4: star formation rate per redshift
plt.scatter(data['sfr'], data['redshift'], marker='.', color='mediumvioletred')
plt.xlabel('Star Formation Rate')
plt.ylabel("Redshift")
plt.title('Star Formation Rate per Redshift')
plt.show()
#zoomed in
plt.scatter(data['sfr'], data['redshift'], marker='.', color='mediumvioletred')
plt.axis([0,20,0,12])
plt.xlabel('Star Formation Rate')
plt.ylabel("Redshift")
plt.title('Star Formation Rate per Redshift: Zoomed In')
plt.savefig("CTA200.sfrperz.pdf")
plt.show()
```

```python
#Question 4: Connected Halo Plot 
#plot
plt.scatter(data['redshift'], data['stellarMass'], marker='.', color='dodgerblue')
#plot the halo ID with descendant ID
plt.plot(data['redshift'],data['stellarMass'], color='lightgreen')
plt.title('Connected Halos')
plt.xlabel("Redshift")
plt.ylabel("Stellar Mass")
plt.savefig("CTA200.Galaxy.Tree.pdf")
plt.show()
```
